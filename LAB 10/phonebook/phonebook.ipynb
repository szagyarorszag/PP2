### Inserting data

```python
import psycopg2 as sql
import csv


insert_data = """
    INSERT INTO phonebook VALUES(DEFAULT, %s, %s);
"""


# connection to database 
connection=sql.connect(host = "localhost",
                       dbname = "suppliers", 
                       user = "postgres", 
                       password = "Bakitzhan566*566",
                       port = 5432 ) 

# it gives us access to database and we can merge data frome database
cursor = connection.cursor() 

def File():
    with open("data.csv", "r") as file: 
        data = csv.reader(file,delimiter=',') 
        for row in data:
            cursor.execute(insert_data , row )

def Console():
    name = input(" Pease enter the name :") 
    phone = input (" Please enter the phone :") 
    cursor.execute(insert_data,(name,phone)) 

method = input('Do you want insert data from console or file ?')

if method == 'file' : File()
elif method == 'console' : Console() 
else : print('Wrong method . Try again')

connection.commit()

cursor.close()

connection.close
```

### Querying data

```python
import psycopg2 as sql 
import pandas as pd 
all_table = '''
    SELECT * FROM phonebook'''
specified_by_id = '''
    SELECT * FROM phonebook WHERE id = %s'''
specified_by_name = '''
    SELECT * FROM phonebook WHERE name = %s'''
specified_by_phone = '''
    SELECT * FROM phonebook WHERE phone = %s'''

# connection to database 
connection=sql.connect(host = "localhost",
                       dbname = "suppliers", 
                       user = "postgres", 
                       password = "Bakitzhan566*566",
                       port = 5432 ) 
# it gives us access to database and we can merge data frome database
cursor = connection.cursor() 

method = input('Please choose by what kind of quering do you want to get a data[all/id/name/phone]') 
if method == 'all' : 
    cursor.execute(all_table)
    data = [row for row in cursor.fetchall()]
    Dataframe = pd.DataFrame(data)
    Dataframe = Dataframe.set_index(0)
    Dataframe = Dataframe.rename(columns={ 1: 'Name' , 2 : 'Phone' })

elif method == 'id' :
    id = input('Input the id :')
    cursor.execute(specified_by_id,(id,))
    data = [row for row in cursor.fetchall()]
    Dataframe = pd.DataFrame(data)
    Dataframe = Dataframe.set_index(0)
    Dataframe = Dataframe.rename(columns={ 1: 'Name' , 2 : 'Phone' })

elif method == 'name' :
    name = input('Input the name :')
    cursor.execute(specified_by_name,(name,))
    data = [row for row in cursor.fetchall()]
    Dataframe = pd.DataFrame(data)
    Dataframe = Dataframe.set_index(0)
    Dataframe = Dataframe.rename(columns={ 1: 'Name' , 2 : 'Phone' })

elif method == 'phone' :
    phone = input('Input the phone :')
    cursor.execute(specified_by_phone,(phone,))
    data = [row for row in cursor.fetchall()]
    Dataframe = pd.DataFrame(data)
    Dataframe = Dataframe.set_index(0)
    Dataframe = Dataframe.rename(columns={ 1: 'Name' , 2 : 'Phone' })

else : print("Wrong method . Try again")

connection.commit()

cursor.close()

connection.close
Dataframe.head(20)
```
